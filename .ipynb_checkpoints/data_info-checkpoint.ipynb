# 라벨링된 이미지 만들어보기

```python
from collections import defaultdict
import numpy as np
import shutil
from tqdm import tqdm
import os
import base64
import cv2
from glob import glob
import random
import matplotlib.pyplot as plt

def get_colors(classes):
    return {c: tuple(map(int, np.random.randint(0, 255, 3))) for c in classes}


def draw_bbox(json_path, save_path, n_images=10):   
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    else:
        shutil.rmtree(save_path)
        os.makedirs(save_path)
       
    categories = {
        '01_ulcer': 1,
        '02_mass': 2,
        '04_lymph': 3,
        '05_bleeding': 4
    }
    
    # 각 class마다 고유한 색상을 부여함
    colors = get_colors(categories.values())
    
    json_list = os.listdir(json_path)
    # 총 n_images개의 이미지에 대하여 라벨링된 bbox를 그린 이미지를 만들어 저장함
    for file_name in tqdm(json_list[:n_images]):
        with open(os.path.join(json_path, file_name), 'r') as f:
            json_file = json.load(f)
        image = np.frombuffer(base64.b64decode(json_file['imageData']), np.uint8)
        image = cv2.imdecode(image, cv2.IMREAD_COLOR)
        
        for shape in json_file['shapes']:
            label = shape['label']
            points = shape['points']
            x1, y1, x2, y2 = round(points[0][0]), round(points[0][1]), round(points[2][0]), round(points[2][1])
            
            cv2.rectangle(image, (x1, y1), (x2, y2), colors[categories[label]], 2)
            (tw, th), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_COMPLEX, 0.6, 1)
            cv2.rectangle(image, (x1, y1-20), (x1+tw, y1), colors[categories[label]], -1)
            cv2.putText(image, label, (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

        file_name = file_name.split('.')[0] + '.jpg'
        cv2.imwrite(os.path.join(save_path, file_name), image)
```

```python
draw_bbox('./data/train/', './examples/', 10)
```

```python
def show_images(images):
    fix, axes = plt.subplots(ncols=len(images), squeeze=False, dpi=200)
    for i, image_path in enumerate(images):
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        axes[0, i].imshow(image)
        axes[0, i].axis('off')


images = glob('./examples/*')
random.shuffle(images)
images = images[:5]

show_images(images)
```
